# MCMC: Gibbs Sampling

```python
n = 500

J = 0.5
B = 0.001
beta = 10

n_steps = 100
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10, 5))
ax0.set_title(f'Step {0}')
ax0.axis('off')
ax1.set_title(f'Step {0}')
ax1.axis('off')

lattice = create_lattice(n, n)
ax0.imshow(lattice, 'binary')

magnetizations = [magnetization(lattice)]
energies = [energy(lattice, J, B)]

for i in range(n_steps):
    step_all(lattice, J, B, beta)
    
    magnetizations.append(magnetization(lattice))
    energies.append(energy(lattice, J, B))
    
    if i % 5 == 0:
        im = ax1.imshow(lattice, 'binary')
        ax1.set_title(f'Step {i}')
        display.display(plt.gcf())
        display.clear_output(wait=True)
```

```python
fig, (ax0, ax1) = plt.subplots(2, 1, sharex=True)
ax0.set_title(f'Energy')
ax1.set_title(f'Magnetization')

ax0.plot(energies)
ax1.plot(magnetizations)

plt.xlabel('Step')

plt.tight_layout();
```
